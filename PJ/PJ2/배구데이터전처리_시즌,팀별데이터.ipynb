```python
import pandas as pd

df = pd.read_csv("./volleyball_raw.csv", encoding='utf8')

df
```

```python
경기별_수치 = df.groupby(["경기번호", "팀명", "결과"]).sum()
경기별_수치
```

```python
경기별_수치=경기별_수치.reset_index().reset_index()
경기별_수치
```

```python
경기별_수치["홈_어웨이"]=경기별_수치['index'].apply(lambda x : 0 if x%2==0 else 1)
경기별_수치
```

```python
tmp = 경기별_수치.copy()
home_index = 경기별_수치[경기별_수치['index']/2+1==경기별_수치['경기번호']].index
away_index = 경기별_수치[경기별_수치['index']/2+1!=경기별_수치['경기번호']].index
경기별_수치.iloc[home_index,-1]=tmp.iloc[away_index,2]
경기별_수치.iloc[away_index,-1]=tmp.iloc[home_index,2]
경기별_수치
```

```python
a = ['GS칼텍스', '흥국생명', '한국도로공사', "IBK기업은행", "KGC인삼공사", "현대건설", "페퍼저축은행"]
b=list(range(0,7))
dic = dict(zip(a,b))
dic
```

# 시즌별 경기 나누기

```python
# 시즌별 경기 나누기
# ex) df17 : 17년도 시즌 경기
df = pd.read_csv('여자배구_경기별_수치_.csv')
df17 = df.loc[df['경기날짜'].between('2017-10-01', '2018-04-30')]
df18 = df.loc[df['경기날짜'].between('2018-10-01', '2019-04-30')]
df19 = df.loc[df['경기날짜'].between('2019-10-01', '2020-04-30')]
df20 = df.loc[df['경기날짜'].between('2020-10-01', '2021-05-30')]
df21 = df.loc[df['경기날짜'].between('2021-10-01', '2022-03-30')]
```

```python
홈_승_5세트_17시즌 = df17[(df17['5SET']>0) & (df17['결과']==1) & (df17['홈_어웨이']==1)]
홈_승_4세트_17시즌 = df17[(df17['4SET']>0) & (df17['5SET'] == 0) &(df17['결과']==1) & (df17['홈_어웨이']==1)]
홈_승_3세트_17시즌 = df17[(df17['4SET']==0) &(df17['결과']==1) & (df17['홈_어웨이']==0)]
어웨이_승_17시즌 =df17[(df17['결과']==1) & (df17['홈_어웨이']==1)]
홈_승_3세트_17시즌['팀명'].value_counts()
```

```python
# 17년도 데이터 처리

# 총경기수 추출
총경기수 = df17.groupby(['경기번호','팀명'])['팀명'].unique().value_counts()
총경기수.reset_index()
경기_17 = pd.DataFrame({'팀명': 총경기수.index, '총경기수':총경기수.values})
경기_17['팀명'] = 경기_17.loc[:,'팀명'].str[0]

# 이긴결과 추출
이긴 = df17.groupby(['경기번호','팀명'])[['결과']].sum()
이긴[이긴['결과'] > 0] = 1
이긴경기 = 이긴.groupby('팀명').sum()
이긴경기.reset_index()

경기_17 = pd.merge(경기_17, 이긴경기, left_on='팀명', right_on='팀명')
경기_17.rename(columns={'결과':'이긴경기'}, inplace=True) 
경기_17 = 경기_17.astype({'총경기수':'int', '이긴경기':'int'})
경기_17['승률'] = (경기_17['이긴경기']/경기_17['총경기수'])*100
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['공격종합_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['오픈_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['시간차_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['이동_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['후위_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['속공_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['퀵오픈_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['서브_성공률']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['디그_성공']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['세트_성공']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['리시브_정확']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17 = pd.merge(경기_17,df17.groupby(['팀명'])[['블로킹_성공']].mean().reset_index(), left_on='팀명', right_on='팀명' )
경기_17['시즌'] = '17-18'
경기_17
```
