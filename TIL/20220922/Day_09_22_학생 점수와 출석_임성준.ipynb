```python
import pandas as pd
```

```python
data = pd.read_excel('student_info.xls')
data
```

```python
data2 = data.copy()
# 거리계산을 위해 정규화 전처리 실행
from sklearn.preprocessing import MinMaxScaler
data2[['Mark','Attended']] = MinMaxScaler().fit_transform(data2[['Mark','Attended']])
data2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수 테스트
for k in range(2,8):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(data2[["Attended","Mark"]])
    # 2행 3열 그래프
    plt.subplot(2,3,k-1) #그래프 순번은 1부터 시작
    plt.tight_layout()
    plt.title(f'k={k}')
    plt.xlabel('Attended')
    plt.ylabel('Mark')
    plt.scatter( 
        data["Attended"], 
        data["Mark"],
        c = pred # color : 같은 값끼리 같은 색을 가짐
    ) 
```

```python
# 학생별 점수와 출석수의 평균
data_grouped = data.groupby('Name').mean()
data_grouped
```

```python
# 거리계산을 위해 정규화 전처리 실행
from sklearn.preprocessing import MinMaxScaler
data_grouped[['Mark','Attended']] = MinMaxScaler().fit_transform(data_grouped[['Mark','Attended']])
data_grouped
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수 테스트
for k in range(2,8):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(data_grouped[["Attended","Mark"]])
    # 2행 3열 그래프
    plt.subplot(2,3,k-1) #그래프 순번은 1부터 시작
    plt.tight_layout()
    plt.title(f'k={k}')
    plt.xlabel('Attended')
    plt.ylabel('Mark')
    plt.scatter( 
        data_grouped["Attended"], 
        data_grouped["Mark"],
        c = pred # color : 같은 값끼리 같은 색을 가짐
    )
# 각 데이터 좌표에 맞는 문자열 출력
    for index,mark,attended in data_grouped.itertuples():
        plt.annotate(f'{index}',(attended+.02,mark+.01))
        plt.tight_layout()
```
