## 2. 동적페이지 : 소스보기와 페이지가 다른 내용 가짐, 자바스크립트가 태그를 생성.

    - 원하는 브라우저용 드라이버 설치 : 크롬, 파이어폭스, 엣지

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%99%98%EC%9C%A8'
res = requests.get(url).text
bs = BeautifulSoup(res,'html.parser')
hwan = bs.select('table.rate_table_info > tbody > tr')
hwan_dict = {}
for i in hwan:
    nation = i.select_one('th').get_text()
    dollar = i.select_one('td').get_text()
    hwan_dict[nation] = dollar
hwan_dict
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%99%98%EC%9C%A8'
res = requests.get(url).text
bs = BeautifulSoup(res,'html.parser')
hwan = bs.find('table',{'class':'rate_table_info'})
hwan = hwan.find('tbody')
hwan = hwan.find_all('tr')
for i in hwan:
    print(i.td.get_text())
```

```python
def get_exchange_rate():
    URL_NAVER_EXCHANGE_RATE = "https://search.naver.com/search.naver?ie=UTF-8&query=%ED%99%98%EC%9C%A8&sm=chr_hty"
    response = requests.get(URL_NAVER_EXCHANGE_RATE)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")
    soup = soup.find("table", attrs={"class" : "rate_table_info"})
    soup = soup.find("tbody")

    exchange_rates = dict()
    for row in soup.find_all("tr"):
        exchange_rates[row.th.get_text()] = row.td.get_text()
    print(type(soup.find_all("tr")))

    return exchange_rates

get_exchange_rate()
```

```python
# 복사 -> 링크주소복사
url = 'https://www.koreabaseball.com/ws/Schedule.asmx/GetScheduleList'
params = {"leId": "1",
"srIdList": "0,9",
"seasonId": "2022",
"gameMonth": "07",
"teamId": ""}
sched_json = requests.post(url,params).json()
sched_json
```

```python
print(type(sched_json['rows']))
print(len(sched_json['rows']))
print(type(sched_json['rows'][0]))
print(type(sched_json['rows'][0]['row']))
print(len(sched_json['rows'][0]['row']))
print(sched_json['rows'][0]['row'][0])
print(sched_json['rows'][0]['row'][1])
print(sched_json['rows'][0]['row'][2])
print(sched_json['rows'][0]['row'][3])
print(sched_json['rows'][0]['row'][4])
print(sched_json['rows'][0]['row'][5])
print(sched_json['rows'][0]['row'][6])
print(sched_json['rows'][0]['row'][7])
print(sched_json['rows'][0]['row'][8])
# 위 데이터를 사전의 리스트로 변환
# 출력예 : [
# {"일시":"07.01(금)","시간":"18:30", "팀": "롯데:LG","점수":"1:2", "장소":"잠실"}
# {"일시":"07.01(금)","시간":"18:30", "팀": "롯데:LG","점수":"1:2", "장소":"잠실"}
# {"일시":"07.01(금)","시간":"18:30", "팀": "롯데:LG","점수":"1:2", "장소":"잠실"}
# ]
```

```python
sched_json = sched_json['rows']
```

```python
sched_json[0]['row'][2]
```

```python
import re   

sched_list = []
for i in sched_json['rows']:
    sched_dict = dict()
    tag_ptn = re.compile("<.*?>") # HTML 태그에 해당하는 패턴
    length = len(i['row'])
    game = re.sub(tag_ptn, ' ', i['row'][length-7]['Text']).split()
    if len(i['row']) == 9:
        date = i['row'][0]['Text']
    sched_dict['일시'] = re.sub(tag_ptn, " ", date)
    sched_dict['시간'] = re.sub(tag_ptn, "", i['row'][length-8]['Text'])
    sched_dict['팀'] = re.sub(tag_ptn, ' ',i['row'][length-7]['Text'])
    if len(game) == 5:
        sched_dict['점수'] = f"{game[1]}:{game[3]}"
    else:
        sched_dict['경기취소'] = i['row'][-1]['Text']
    sched_dict['장소'] = i['row'][length-2]['Text']
    sched_list.append(sched_dict)

sched_list
```
